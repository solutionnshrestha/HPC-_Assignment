```python
import matplotlib.pyplot as plot
import pandas as pd
```

```python
line=pd.read_csv('lineara.csv')
line.head()
```

```python
plot.figure(figsize=(10,8))
plot.scatter(line.x,line.y,color='brown')
plot.title("Scatter plot of given dataset",fontsize='14',color='brown')
plot.grid(color='gray')
plot.xlabel('X-axis',fontsize='14')
plot.ylabel('Y-axis',fontsize='14')
plot.show()
```

```python
original=pd.read_csv('original.csv')
original.head()
```

```python
plot.figure(figsize=(10,8))
plot.scatter(line.x,line.y,color='brown')
plot.plot(original.X,original.Y,c='black',linewidth=2, markersize=12)

plot.title("Best Solution",fontsize='14',color='black')
plot.grid()
plot.xlabel('X-axis',fontsize='14')
plot.ylabel('Y-axis',fontsize='14')
plot.show()
```

```python
guess1=pd.read_csv('guess1.csv')
guess1.head()
```

```python
plot.figure(figsize=(10,8))
plot.scatter(line.x,line.y,color='brown')
plot.plot(guess1.X,guess1.Y,c='r',linewidth=2, markersize=12)

plot.title("Guess One",fontsize='14',color='r')
plot.grid()
plot.xlabel('X-axis',fontsize='14')
plot.ylabel('Y-axis',fontsize='14')
plot.show()
```

```python
plot.figure(figsize=(10,8))
plot.scatter(line.x,line.y,color='brown')
line3=plot.plot(guess3.X,guess3.Y,c='purple',linewidth='2', markersize='12')
line2=plot.plot(guess2.X,guess2.Y,c='b',linewidth='2', markersize='12')
line1=plot.plot(guess1.X,guess1.Y,c='r',linewidth='2', markersize='12')

plot.grid()
plot.title('All three guesses',fontsize='14')
plot.legend()
plot.show()
```
